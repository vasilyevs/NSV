### Нормальное распределение

```python
#среднее = 4, стандартное отклонение = 1
sample = sps.norm(loc=4, scale=1)
sns.histplot(data=sample.rvs(size=200))
```

```python
#значение функции распределения в точке 5
sample.cdf(5)
```

```python
x = np.linspace(1,6,100)
cdf = sample.cdf(x)
plt.plot(x, cdf)
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.show
plt.grid()
```

```python
x = np.linspace(1,6,100)
pdf = sample.pdf(x) # значение плотности в точке
plt.plot(x, pdf)

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show
plt.grid()
```

Известно, что светофор горит 54 секунды красным, 4 секунды жёлтым и 28 секунд зелёным. Посчитайте, какое количество информации несёт сообщение о цвете светофора в текущий момент. А что будет в случае 28, 28 и 28 секунд для каждого цвета?

```python
# первый вариант распределения времени. формула Шеннона
i1 = round(-((54/86)*math.log2(54/86)+(4/86)*math.log2(4/86)+(28/86)*math.log2(28/86)), 3)
# вторй вариант распределения времени. формула 
i2 = round(math.log2(1/(1/3)), 3)
print('I1', i1, 'бит')
print('I2', i2, 'бит')
```
